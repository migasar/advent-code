### Format Input

```python
import numpy as np
from itertools import product

m = np.array([[int(x) for x in list(col)] for col in puzzle_input.splitlines()])
#m
```

## Part One

```python
low_values = [m[r, c] for r,c in product(*map(range, m.shape)) if m[r,c] == m[max(0,r-1):r+2,max(0,c-1):c+2].min()]
```

```python
result = sum(np.array(low_values) + 1)
result
```

## Part Two 

```python
low_points = [[r, c] for r,c in product(*map(range, m.shape)) if m[r,c] == m[max(0,r-1):r+2,max(0,c-1):c+2].min()]
#low_points
```

```python
def measure_basin(r, c, zone=[]):

    if m[r, c] < 9 and [r, c] not in zone:
        zone.append([r, c])
        
        adjacents = [
            [min(r+1, (len(m)-1)), c],
            [max(0,r-1), c],
            [r, min(c+1, (len(m[0])-1))],
            [r, max(0,c-1)]
        ]
        
        for position in adjacents:
            measure_basin(position[0], position[1], zone)

    return zone
```

```python
sizes = []
for point in low_points:
    basin = measure_basin(point[0], point[1], zone=[])
    sizes.append(len(basin))
```

```python
result = np.prod(sorted(sizes, reverse= True)[:3])
result
```
