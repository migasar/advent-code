### Set Input

```python
# Define inputs
puzzle_input_1 = INPUT_FULL_1
puzzle_input_2 = INPUT_FULL_2
```

```python
# Format first input

import numpy as np


def parse_dots(puzzle_input):
    """Parse input"""
    data = np.array([
        [int(x) for x in line.split(',')] for line in puzzle_input.splitlines()
        ])
    return data

dots = parse_dots(puzzle_input_1)
#dots
```

```python
# Format second input

def parse_folds(puzzle_input):
    """Parse input"""
    
    data = []
    pattern = "fold along "

    for line in puzzle_input.splitlines():
        match = line.lstrip(pattern)
        instruction = match.split('=')
        instruction[1] = int(instruction[1])
        data.append(instruction)

    return data

folds = parse_folds(puzzle_input_2)
#folds
```

## Part 1

### Question

### Solution

```python
# Find shape of board
board = np.zeros((dots[:, 1].max() + 1, dots[:, 0].max() + 1), dtype=int)
board.shape
```

```python
# Fill board with dots
for x,y in dots:
    board[(y, x)] = 1
board
```

```python
# Split the board 

# Find the folder line
fold = folds[0][1]
print(f"{fold=}")

# Fold on horizontal line
if folds[0][0] == 'y':
    board_up = board[:fold].copy()
    board_down = board[fold+1:].copy()
    new_board = board_up + np.flip(board_down, 0)

# Fold on vertical line:
else:
    board_left = board[:, fold:].copy()
    board_right= board[:, :fold+1].copy()
    new_board = board_left + np.flip(board_right, 1)

print(new_board)
```

```python
# Counting number of dots in folded board
count = np.count_nonzero(new_board)
count
```

```python
# Split the board 

for axis in folds:
    # Find the folder line
    print(f"{axis=}")
    fold = axis[1]
    #print(f"{fold=}")

    # Fold on horizontal line
    if axis[0] == 'y':
        #board_up = board[:fold+1].copy()
        board_up = board[:fold:,:]
        #board_down = board[fold:].copy()
        board_down = board[-fold::,:]
        board = board_up + np.flip(board_down, 0)

    # Fold on vertical line:
    else:
        #board_left = board[:, fold+1:].copy()
        board_left = board[:, :fold:]
        #print("board_left:")
        #print(board_left)
        #board_right= board[:, :fold].copy()
        board_right= board[:, -fold::]
        #print("board_right:")
        #print(board_right)
        board = board_left + np.flip(board_right, 1)

    print("new board:")
    print(board)

    # Counting number of dots in folded board
    count = np.count_nonzero(board)
    print(f"{count=}")
```
